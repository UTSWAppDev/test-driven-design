```python
import unittest
from unittest import TestCase, TestSuite, TextTestRunner
from unittest.mock import patch, Mock, MagicMock
import sys
from io import StringIO

print("✓ All required libraries imported successfully")
```

```python
def isEven(n):
    if n % 2 == 0:
        return True
    else:
        return False

class TestIsEven(unittest.TestCase):
    '''Tests for isEven function'''
    
    def test_is_even_with_even_number(self):
        '''Does an even number return true?'''
        self.assertTrue(isEven(4))

    def test_is_even_with_odd_number(self):
        '''Does an odd number return false?'''
        self.assertFalse(isEven(5))

print("✓ Test case class defined successfully")
```

```python
if __name__ == '__main__':
    # Create a test suite
    suite = unittest.TestLoader().loadTestsFromTestCase(TestIsEven)
    
    # Run with verbose output
    runner = unittest.TextTestRunner(verbosity=2)
    result = runner.run(suite)
    
    # Print summary
    #print(f"\n{'='*50}")
    print(f"Tests run: {result.testsRun}")
    print(f"Successes: {result.testsRun - len(result.failures) - len(result.errors)}")
    print(f"Failures: {len(result.failures)}")
    print(f"Errors: {len(result.errors)}")
    print(f"{'='*50}")
```
