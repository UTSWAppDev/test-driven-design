## 2. Creating our first Test-Driven Function

### What is TDD?
Test-Driven Design is a software development methodology that emphasizes writing tests **before** writing the actual implementation code.

### **STEPS**
1. Document the function's purpose. Use plain language here.
2. Define and document the function's _ins_ and _outs_
   1. _Ins_ are the parameter types that the function expects.
   2. _Outs_ is the type of the return value.
3. Create a placeholder function. The function doesn't have to work yet, it just needs to be available for the test suite to see.
4. Build the tests
5. Check that the tests run. Most should fail, since they're running on a placeholder function at this point.
5. Go back and write the real function to replace the placeholder
6. Ensure that all tests pass

---

### Benefits of TDD
- **Bug Prevention**: Catch bugs early before they enter production
- **Better Design**: Forces you to think about the interface and requirements upfront
- **Documentation**: Tests serve as living documentation of what the code does
- **Confidence**: Comprehensive test coverage gives confidence when refactoring
- **Fewer Defects**: Studies show TDD reduces defects by 40-80%

### Steps 1-3

Run the following code block to complete steps 1-3.

````python
# Purpose: isEven() checks if a number is even. It returns True if the number is even, and False otherwise. # <-- Step 1 Function Purpose
# IsEven: int -> bool                               # <-- Step 2 Ins and Outs
def isEven(n):                                      # <-- Step 3 Function Placeholder
    return True                                     # <-- Step 3 This will replaced by actual implementation later

print("✓ Function isEven() placeholder defined successfully. Steps 1-3 complete.")
````

### Step 4 Creating the Tests
#### The unittest Module
We will use Python's `unittest` module. It is part of the standard library and provides a structured way to create and execute tests:

The basic requirements for creating a test case:
- Create a test class that inherits from `unittest.TestCase`
  ```python
  class TestIsEven(unittest.TestCase):
    '''Tests for isEven function'''
  ```
- Create at least one test by defining the test function and providing an `assertion`. Note the common assertion methods below
  ```python
  def test_iseven_with_even_number(self):
        '''Does an even number return true?'''
        self.assertTrue(isEven(4))
  ```
- You can create as many tests as you like using the same pattern.    

### Basic Structure
A test class must:
1. Inherit from `unittest.TestCase`
2. Define test methods starting with `test_`
3. Use assertion methods to verify behavior

### Common Assertion Methods
| Method | Purpose |
|--------|---------|
| `assertEqual(a, b)` | Check if a == b |
| `assertNotEqual(a, b)` | Check if a != b |
| `assertTrue(x)` | Check if x is True |
| `assertFalse(x)` | Check if x is False |
| `assertIn(a, b)` | Check if a is in b |
| `assertNotIn(a, b)` | Check if a is not in b |
| `assertIsNone(x)` | Check if x is None |
| `assertIsNotNone(x)` | Check if x is not None |


---

Next, see an example of the test class creation. Run the following code block to create the tests for step 4. They won't execute yet, they'll just be defined.


````python
class TestIsEven(unittest.TestCase):             # <- Step 4 Test Case Class
    '''Tests for isEven function'''
    
    def test_iseven_with_even_number(self):     # Define a test method
        '''Does an even number return true?'''  # This is the docstring for the test so that we see the output clearly when it runs
        self.assertTrue(isEven(4))              # Notice the use of assertTrue. This checks if the result of isEven(4) is True.

    def test_iseven_with_odd_number(self):      # Define another test method
        '''Does an odd number return false?'''
        self.assertFalse(isEven(5))             # Notice the use of assertFalse. This checks if the result of isEven(5) is False.

print("✓ Test case class defined successfully. Step 4 complete.")
````

### Step 5 Validate the Tests and Placeholder Functions
In this step, we will validate that our tests and placeholder functions are synced up.

If our placeholder function doesn't exist, the tests will not run.

If our placeholder function does exist, the tests should run **and many should fail**

In this step, **failure is success**!

Run the following code block to execute the tests against the placeholder function:

````python
if __name__ == '__main__':
    # Create a test suite
    suite = unittest.TestLoader().loadTestsFromTestCase(TestIsEven)     # here we load our TestIsEven test case class into the test suite
    
    # Run with verbose output
    runner = unittest.TextTestRunner(verbosity=2)
    result = runner.run(suite)    # this runs all the tests in the suite and stores the results in 'result'
    
    # Print summary
    # Don't worry about this part for now; it's just to give a pretty summary of the test results
    print(f"\n{'='*50}")
    print(f"Tests run: {result.testsRun}")
    print(f"Successes: {result.testsRun - len(result.failures) - len(result.errors)}")
    print(f"Failures: {len(result.failures)}")
    print(f"Errors: {len(result.errors)}")
    print(f"{'='*50}")

    print("✓ Tests executed successfully. Step 5 complete.")
````

Look above at the test that failed.

`test_iseven_with_odd_number` failed.

This is because our placeholder function returns True regardless of the input, so it will be correct for even numbers and incorrect for odd numbers.

This is success, it means that our tests are working.

### Step 6 Replace the Placeholder Function With The Real Thing
In this step, we will refactor our placeholder function with real logic.

Execute the code block below to replace the placeholder function with the real one:

````python
# Purpose: isEven() checks if a number is even. It returns True if the number is even, and False otherwise. # <-- Step 1 Function Purpose
# IsEven: int -> bool                               # <-- Step 2 Ins and Outs
# Now we will implement the actual function for step 6
def isEven(n):
    if n % 2 == 0:  
        return True
    else:
        return False                                      

print("✓ Function isEven() implemented successfully. Step 6 complete.")
````

### Step 7 Validate the Function Implementation
This is the last step of the process, where it all comes together.

First we defined the function and its ins and outs.

Then we created a placeholder function so that we could create our tests.

We created the tests and validated that they worked against the placeholder function.

We rewrote the function with actual logic.

Now it's time to test our function to see if we wrote it successfully.

Run the following code block to test your function implentation. You may notice that it is the exact same code that we ran on the placeholder function.

````python
if __name__ == '__main__':
    # Create a test suite
    suite = unittest.TestLoader().loadTestsFromTestCase(TestIsEven)     # here we load our TestIsEven test case class into the test suite
    
    # Run with verbose output
    runner = unittest.TextTestRunner(verbosity=2)
    result = runner.run(suite)    # this runs all the tests in the suite and stores the results in 'result'
    
    # Print summary
    # Don't worry about this part for now; it's just to give a pretty summary of the test results
    print(f"\n{'='*50}")
    print(f"Tests run: {result.testsRun}")
    print(f"Successes: {result.testsRun - len(result.failures) - len(result.errors)}")
    print(f"Failures: {len(result.failures)}")
    print(f"Errors: {len(result.errors)}")
    print(f"{'='*50}")

    print("✓ Tests executed successfully. Step 7 complete.")
````

**Hooray!**
Now we have two tests that pass. This lets us know that our function works for both even and odd numbers, so we can safely put it into production!

Congratulatons, you just wrote your first Test Driven Function!

## Key Takeaways: TDD Best Practices

### ✓ DO
1. **Write tests first**: Tests define requirements
2. **Write minimal code**: Only enough to pass tests
3. **Test one thing**: Each test should verify one behavior
4. **Use descriptive names**: Test method names should explain what they test
5. **Keep tests independent**: Tests should not depend on each other
6. **Use assertions liberally**: Verify all important behaviors
7. **Test edge cases**: Empty strings, None, zero, negative numbers, etc.
8. **Refactor with confidence**: Tests catch regressions

### ✗ DON'T
1. **Skip testing** "obvious" code: Bugs hide everywhere
2. **Test implementation details**: Test behavior, not internals
3. **Create complex tests**: Simple, focused tests are better
4. **Ignore test failures**: Fix them immediately
5. **Test too much in one test**: Keep assertions focused


### TDD Benefits Summary
| Benefit | Impact |
|---------|--------|
| Bug Prevention | Catch bugs before production |
| Design Quality | Better code structure |
| Documentation | Tests show how to use code |
| Refactoring Safety | Change code confidently |
| Debugging Speed | Know exactly what's broken |
| Code Confidence | High test coverage = peace of mind |

## Appendix: Running Tests from the Command Line

### Running All Tests in a File
```bash
python -m unittest test_mymodule.py
```

### Running Specific Test Class
```bash
python -m unittest test_mymodule.TestClassName
```

### Running Specific Test Method
```bash
python -m unittest test_mymodule.TestClassName.test_method_name
```

### Verbose Output
```bash
python -m unittest test_mymodule.py -v
```

### Discover and Run All Tests
```bash
python -m unittest discover -s . -p "test_*.py" -v
```
